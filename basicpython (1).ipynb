```python
# 13. Write a Python program to find the sum of all even numbers between 1 and 50.
total_sum = 0

for i in range(1, 51):
    if i % 2 == 0:
        total_sum += i

print("Sum of even numbers:", total_sum)
```

```python
# 14. Write a program to reverse a string using a while loop.
text = "Python"
reversed_text = ""
index = len(text) - 1

while index >= 0:
    reversed_text += text[index]
    index -= 1

print("Original:", text)
print("Reversed:", reversed_text)
```

```python
# 15. Write a Python program to calculate the factorial of a number provided by the user using a while loop.
num = int(input("Enter a number: "))
factorial = 1
count = num

while count > 0:
    factorial *= count
    count -= 1

print(f"Factorial of {num} is {factorial}")
```

1. What is Python, and why is it popular?
 --Python is a high-level programming language that is very easy to read and write because its syntax looks a lot like simple English. It is popular because it is versatile; you can use it for almost anything, including web development, data science, artificial intelligence, and automation. It also has a huge community, meaning there are endless free libraries and tools available to help you build things faster.

2. What is an interpreter in Python?  
--An interpreter is like a translator that works line-by-line. Instead of translating your entire code into machine language at once (like a compiler), the Python interpreter reads one line of your code, translates it, executes it, and then moves to the next. This makes it easier to find errors because the program stops exactly where the mistake happened.

3. What are pre-defined keywords in Python?
 --Keywords are special reserved words that Python has already claimed for its own internal use. Each keyword has a specific meaning and triggers a specific action. Examples include if, else, while, for, def, and import.

4. Can keywords be used as variable names?
-- No, you cannot use keywords as variable names. Since Python uses these words to understand the structure of your program (like deciding when to loop or define a function), using them as variable names would confuse the interpreter and cause a syntax error.

5. What is mutability in Python?
 --Mutability refers to the ability to change an object after it has been created. If an object is mutable, you can modify its internal data (like adding items to a list) without creating a completely new object. If it is immutable, you cannot change it once it is made; you would have to create a new one to store a different value.

6. Why are lists mutable, but tuples are immutable?
 --Lists are designed to be dynamic collections where you might need to add, remove, or change items frequently, so they are made mutable. Tuples are designed to be "write-protected" collections for data that shouldn't change, like coordinates or configuration settings. This immutability makes tuples slightly faster and safer for storing constant data.

7. What is the difference between "==" and "is" operators in Python?  
--The == operator checks for equality of value (do these two things look the same?). The is operator checks for identity (are these two things actually the exact same object in memory?).
a == b: True if a and b have the same contents.
a is b: True if a and b point to the same memory address.

8. What are logical operators in Python?  
--Logical operators are used to combine conditional statements.
and: Returns True if both statements are true.
or: Returns True if at least one statement is true.
not: Reverses the result (returns False if the result is true).

9. What is type casting in Python?  
--Type casting is the process of converting a variable from one data type to another. For example, changing the string "123" into the integer 123 so you can perform math on it.

10. What is the difference between implicit and explicit type casting?
--Implicit: Python does it automatically for you. For example, if you add an integer (3) and a float (4.5), Python automatically converts the answer to a float (7.5) to prevent data loss.

--Explicit: You manually force the conversion using functions like int(), str(), or float(). You do this when Python doesn't know how to convert it automatically, like turning a string into a number.

11. What is the purpose of conditional statements in Python?
 --Conditional statements allow your program to make decisions. They let the code execute specific blocks only if certain conditions are met (True). Without them, code would just run straight through from top to bottom every time.

12. How does the elif statement work?  
--elif stands for "else if". It is used to check multiple conditions in a sequence. If the first if condition is False, Python checks the elif. If the elif is True, it runs that block. You can have many elif statements to handle different possibilities one by one.

13. What is the difference between for and while loops?
--For loop: Best used when you know exactly how many times you want to iterate (like looping through a list of 10 items). It runs for a specific set of elements.

--While loop: Best used when you want to loop until a specific condition changes. You don't necessarily know how many times it will run beforehand; it just keeps going as long as the condition is True.

14. Describe a scenario where a while loop is more suitable than a for loop.  
--A while loop is better when you are waiting for an external event or user input. For example, a game menu that keeps asking "Do you want to play again?" requires a while loop because you don't know if the user will say "yes" once or fifty times
